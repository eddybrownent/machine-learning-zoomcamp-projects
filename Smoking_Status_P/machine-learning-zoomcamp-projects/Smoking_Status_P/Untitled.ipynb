```python
import pandas as pd
from scipy import stats


# Get a list of all numerical columns
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Define a threshold for identifying outliers
outlier_threshold = 3  # You can adjust this threshold as needed

# Create an empty list to store outlier dataframes
outliers_df_list = []

# Iterate through numerical columns and check for outliers
for column in numerical_columns:
    # Calculate Z-scores for the current column
    z_scores = stats.zscore(data[column])
    
    # Create a mask to identify outliers
    outliers = (z_scores > outlier_threshold) | (z_scores < -outlier_threshold)
    
    # Get the outlier values for the current column
    outlier_values = data[column][outliers]
    
    # Create a DataFrame with the outlier values
    outlier_df = pd.DataFrame({'Column': [column] * len(outlier_values), 'Outlier Values': outlier_values.tolist()})
    
    # Append the outlier DataFrame to the list
    outliers_df_list.append(outlier_df)

# Concatenate the list of outlier DataFrames into a single DataFrame
outliers_df = pd.concat(outliers_df_list, ignore_index=True)

# Print the DataFrame with outlier values
print(outliers_df)
 
```

```python
import pandas as pd
from scipy import stats

# Load your dataset


# Get a list of all numerical columns
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns

# Define a threshold for identifying outliers
outlier_threshold = 2  # You can adjust this threshold as needed

# Create a dictionary to store outlier values for each column
outlier_values_dict = {}

# Iterate through numerical columns and check for outliers
for column in numerical_columns:
    # Calculate Z-scores for the current column
    z_scores = stats.zscore(data[column])
    
    # Create a mask to identify outliers
    outliers = (z_scores > outlier_threshold) | (z_scores < -outlier_threshold)
    
    # Get the outlier values for the current column
    outlier_values = data[column][outliers]
    
    # Store the outlier values in the dictionary
    outlier_values_dict[column] = outlier_values

# Print the outlier values for each column
for column, values in outlier_values_dict.items():
    print(f"Outlier values in '{column}':")
    print(values)
    print('\n')
```
